```python
caminho_imagem = '/content/condutora.png'

try:
    imagem_original = Image.open(caminho_imagem)
    #display
    imagem_original.show()
    #display Colab
    display(imagem_original)
except FileNotFoundError:
    print(f"Erro: A imagem n√£o foi encontrada em {caminho_imagem}")
except Exception as e:
    print(f"Ocorreu um erro: {e}")
```

```python
#converter uma imagem rgb em escala de cinzas
def grizzy(original_colorida):
    original_colorida_np = np.array(original_colorida)
    r, g, b = original_colorida_np[:, :, 0], original_colorida_np[:, :, 1], original_colorida_np[:, :, 2]
    grayscale_values = (0.299 * r + 0.587 * g + 0.114 * b).astype(np.uint8)
    return grayscale_values
```

```python
imagem_array = np.array(imagem_original)

original_grizzy_temp = grizzy(imagem_array)
```

```python
imagem_grizzy = Image.fromarray(original_grizzy_temp)

imagem_grizzy.save('original_grizzy.png')

original_grizzy = Image.open('original_grizzy.png')

display(original_grizzy)
```

```python
#converter uma imagem rgb ou em escala de cinzas em bin√°ria (preto e branco)

imagem_cinza = Image.open('original_grizzy.png')

limite_rgb = 127

imagem_binaria = imagem_cinza.point(lambda p: 0 if p < limite_rgb else 255, '1')

imagem_binaria.save('original_binaria.png')

original_binaria = Image.open('original_binaria.png')

display(original_binaria)
```
